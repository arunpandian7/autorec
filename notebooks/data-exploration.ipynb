```python
stock_description_pair = {}
for id in df["StockCode"].unique():
    stock_description_pair[id] = df.groupby("StockCode").get_group(id)["Description"].unique()
```

```python
len(stock_description_pair)
```

```python
count = 0
for key, val in stock_description_pair.items():
    if len(val) > 1 :
        count +=1
        print(key, val)
print(f"Number of products having different description:", count)
```
